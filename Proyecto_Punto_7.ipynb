Analisis con Neutral con Label

```python
finance.head()
```

```python
X = finance['Text']
y = finance['Label'].map({'positive':1,'negative':0})
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.23,shuffle=True)
```

```python
vectorizacion = CountVectorizer(stop_words='english',ngram_range=(1,4))
```

```python
X_train_vec = vectorizacion.fit_transform(X_train.values)
X_test_vec = vectorizacion.transform(X_test.values)
```

```python
model = MultinomialNB()
```

```python
model.fit(X_train_vec,y_train)
```

```python
y_pred = model.predict(X_test_vec)
```

```python
print(y_pred)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
Accuracy1 = accuracy_score(y_test,y_pred)
```

```python
print(confusion_matrix(y_test,y_pred))
```
